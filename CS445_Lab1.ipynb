```python
#necessary import statements
import pandas as pd

#customize the column names
column_names = [
    "symboling", "normalized-losses", "make", "fuel-type", "aspiration",
    "num-of-doors", "body-style", "drive-wheels", "engine-location",
    "wheel-base", "length", "width", "height", "curb-weight", "engine-type",
    "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke",
    "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg",
    "price"
]

#read the data in the csv file
df = pd.read_csv('imports-85.data', names=column_names)

#display a preview of the data
print(df.head())
```

```python
#necessary import statements
import pandas as pd
import matplotlib.pyplot as plt

#count number of cars per make
make_counts = df['make'].value_counts()

#create a bar chart using this data
make_counts.plot(kind='bar', color='skyblue')

#customize the labels and title
plt.title('Number of Cars per Make', fontsize=16)
plt.xlabel('Car Make', fontsize=12)
plt.ylabel('Num Cars', fontsize=12)

#display the plot
plt.show()

#save the chart as a png
plt.savefig('FisherBachman-Rhodes_CarBarChart.png')

#
# this bar chart displays the number of cars of each make as as bars
# the count of cars is on the y axis 
# the different car makes is on the x axis
#
```

```python
#necessary import statements
import pandas as pd
import matplotlib.pyplot as plt

#count number of cars per make
make_counts = df['make'].value_counts()

#create a line chart using this data
make_counts.plot(kind='line', color='skyblue', marker='o')

#customize the labels and title
plt.title('Number of Cars per Make', fontsize=16)
plt.xlabel('Car Make', fontsize=12)
plt.ylabel('Count', fontsize=12)

#customize x-axis to show all car makes
plt.xticks(range(len(make_counts)), make_counts.index, rotation=90)

#display the plot
plt.show()

#save the chart as a png
plt.savefig('FisherBachman-Rhodes_CarLineChart.png')

#
# this line chart displays the number of cars of each make as dots. These dots are then connected using lines
# the count of cars is on the y axis 
# the different car makes is on the x axis
#
```

```python
#necessary import statements
import pandas as pd
import matplotlib.pyplot as plt

#count the number of cars per make
make_counts = df['make'].value_counts()

#create a pie chart using this data
make_counts.plot(kind='pie', labels=[f'{make} ({count})' for make, count in make_counts.items()], color='skyblue')

#customize the labels and title
plt.title('Number of Cars per Make', fontsize=16)

#display the plot
plt.show()

#save the chart as a png
plt.savefig('FisherBachman-Rhodes_CarPieChart.png')



#
# this pie chart displays the number of cars using differently sized sections of a "pie" corresponding to their percentage of total cars
# the count of cars is represented with "pie slices" of different sizes
# the different car makes are titled around the chart
#
```

```python
#necessary import statements
import pandas as pd
import matplotlib.pyplot as plt

#count the number of cars per make
make_counts = df['make'].value_counts()

#create a box chart using this data
make_counts.plot(kind='box', color='skyblue')

#customize labels and title of the chart
plt.title('Number of Cars per Make', fontsize=16)
plt.xlabel('Car Make', fontsize=12)
plt.ylabel('Count', fontsize=12)

#display the chart
plt.show()

#save the chart as a png
plt.savefig('FisherBachman-Rhodes_CarBoxChart.png')



#
# this box chart displays the minimum, first quartile, median, third quartile, and maximum make counts
# the count of cars is on the y axis 
#
```
